```python
from transformers import pipeline
```

```python
bert_model = "bert-base-uncased"
roberta_model = "roberta-base"
bart_model = "facebook/bart-base"
```

```python
gen_bert = pipeline("text-generation", model=bert_model)
gen_bert("The future of Artificial Intelligence is", max_length=30)
```

```python
gen_roberta = pipeline("text-generation", model=roberta_model)
gen_roberta("The future of Artificial Intelligence is", max_length=30)
```

```python
gen_bart = pipeline("text-generation", model=bart_model)
gen_bart("The future of Artificial Intelligence is", max_length=30)
```

```python
fill_bert = pipeline("fill-mask", model=bert_model)
fill_bert("The goal of Generative AI is to [MASK] new content.")
```

```python
fill_roberta = pipeline("fill-mask", model=roberta_model)
fill_roberta("The goal of Generative AI is to <mask> new content.")
```

```python
fill_bart = pipeline("fill-mask", model=bart_model)
fill_bart("The goal of Generative AI is to <mask> new content.")
```

```python
context = "Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
question = "What are the risks?"
```

```python
qa_bert = pipeline("question-answering", model=bert_model)
qa_bert(question=question, context=context)
```

```python
qa_roberta = pipeline("question-answering", model=roberta_model)
qa_roberta(question=question, context=context)
```

```python
qa_bart = pipeline("question-answering", model=bart_model)
qa_bart(question=question, context=context)
```

| Task       | Model   | Classification | Observation                      | Architectural Reason             |
| ---------- | ------- | -------------- | -------------------------------- | -------------------------------- |
| Generation | BERT    | Failure        | Could not generate coherent text | Encoder-only, not autoregressive |
|            | RoBERTa | Failure        | Similar failure                  | Encoder-only architecture        |
|            | BART    | Success        | Generated fluent text            | Encoder–Decoder (seq2seq)        |
| Fill-Mask  | BERT    | Success        | Correct word prediction          | Trained with MLM                 |
|            | RoBERTa | Success        | More confident predictions       | Optimized MLM training           |
|            | BART    | Partial        | Less precise                     | Not MLM-focused                  |
| QA         | BERT    | Partial        | Short answers                    | No QA fine-tuning                |
|            | RoBERTa | Partial        | Slight improvement               | Better representations           |
|            | BART    | Failure        | Irrelevant output                | Generative, not extractive       |
